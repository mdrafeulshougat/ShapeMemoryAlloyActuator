```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset1 = pd.read_csv(r'C:\Users\Md Rafeul Shougat\Data_SMA\XGBoost_SMA_x.csv')
dataset2 = pd.read_csv(r'C:\Users\Md Rafeul Shougat\Data_SMA\XGBoost_SMA_y.csv')
x = dataset1.iloc[:,:].values
y = dataset2.iloc[:,:].values
```

```python
plt.rcParams.update({'font.size': 20})
x1=np.linspace(0, 1.0, num=len(x))
plt.rcParams['figure.figsize'] = [20, 5]
plt.plot(x1,x[:,0],'-b',linewidth=2)
plt.plot(x1,x[:,1],'-r',linewidth=1)
#plt.ylim(-10,45)
#plt.ylim(45,95)
plt.xlim(0.8,1.0)
plt.xlabel('samples')
#plt.ylabel('Y axis deformation')
plt.ylabel('X axis location')
plt.legend(['target','prediction'],loc ="lower right")
```

```python
plt.rcParams.update({'font.size': 50})
x1=np.linspace(0, 1.0, num=len(y))
plt.rcParams['figure.figsize'] = [20, 5]
plt.plot(x1,y[:,0],'-b',linewidth=2)
plt.ylabel('samples')
plt.plot(x1,y[:,1],'-r',linewidth=1)
plt.ylim(-10,45)
#plt.ylim(45,95)
plt.xlim(0.8,1.0)
plt.xlabel('samples')
plt.ylabel('Y axis deformation')
#plt.ylabel('X axis location')
plt.legend(['target','prediction'],loc ="lower right")
```

```python
fig, axs = plt.subplots(2)
#fig.suptitle('Vertically stacked subplots')
plt.rcParams.update({'font.size': 20})
plt.rcParams['figure.figsize'] = [20, 8]

axs[0].plot(x1,x[:,0],'-b',linewidth=3)
axs[0].plot(x1,x[:,1],'-r')
axs[0].set_ylabel("X location")
axs[0].set_xlim(0,1)

axs[1].plot(x1,y[:,0],'-b',linewidth=3)
axs[1].plot(x1,y[:,1],'-r')
axs[1].set_ylabel("Y deflection")
axs[1].set_xlim(0,1)
axs[1].set_xlabel("time")

for ax in axs.flat:
    ax.label_outer()
# for ax in axs.flat:
#     ax.set(xlabel='time', ylabel='deflection')

labels = ["target", "prediction"]
fig.legend(labels=labels,bbox_to_anchor=(0.185, 0.5),loc ="center")

plt.savefig('Data_SMA\XGBoost_SMA.png',bbox_inches='tight')
#plt.savefig('RF.png', bbox_inches='tight')
plt.show()
```
